### Visualization

## Barplot between Sub-category vs Category

```python
udf=pd.read_csv('Superstore.csv',index_col='Row ID')
```

```python
udf
```

```python
plt.figure(figsize=(20,8))
plt.bar('Sub-Category','Category', data=udf)
plt.show()
```

## Pie plot od Sub-category

```python
plt.figure(figsize=(10,10))
udf['Sub-Category'].value_counts().plot.pie(autopct="%1.1f%%",  textprops = {'fontsize':10, 'color':'black'})
plt.show()
```

```python
udf.groupby('Sub-Category')[['Profit','Sales']].agg(['sum']).plot.bar()
plt.title("Total profit and sales per sub-Category")
plt.rcParams['figure.figsize']=[10,8]
plt.show()
```

## Count of city Plot

```python
print(udf['State'].value_counts())
plt.figure(figsize=(15,8))
sns.countplot(x=udf['State'])
plt.xticks(rotation=90)
plt.show()
```

## Sales and Profit Region Wise

```python
udf.groupby('Region')[['Profit','Sales']].agg(['sum']).plot.bar()
plt.title("total Profit and Sales per region")
#plt.legend('Profit')
#plt.legend('Sales')
plt.rcparams['figure.figsize']=[10,8]
plt.show()
```

## Count of sub-category

```python
print(udf['Sub-Category'].value_counts())
plt.figure(figsize=(13,6))
sns.countplot(x=udf['Sub-Category'])
plt.xticks(rotation=90)
plt.show()
```
